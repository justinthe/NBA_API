```python
gameIds = list()
```

```python
for gameId in gameIds:
    response = requests.get("https://api-nba-v1.p.rapidapi.com/gameDetails/"+gameId,
                           headers={
                               "X-RapidAPI-Host": RAPIDAPI_HOST,
                               "X-RapidAPI-Key": RAPIDAPI_KEY
                           })
    if(response.status_code == 200):
        games = response.json()
```
